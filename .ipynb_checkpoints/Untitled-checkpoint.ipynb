```python
import pandas as pd
```

## Importing Change Order info
#### Grouped by project

```python
co = pd.read_csv('data/raw_data/change_orders.csv',index_col='CONT_ID')['C_O_AMT']
co.columns = ['co_amt']
co = co.map(lambda x: float(x.replace('$','').replace(',','')))
co = co.groupby(co.index).sum()
```
